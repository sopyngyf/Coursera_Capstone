# Segmenting and Clustering Neighborhoods in Toronto

```python
#pip install lxml done previously hence not shown here.
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
df = pd.read_html(url)
df1 = df[0]
df1.head(10)
```

# Drop rows with a borough that is 'Not assigned'

```python
df2 = df1[df1.Borough !='Not assigned'].reset_index(drop=True)
df2.head(10)
```

# If neighbourhood is not assigned, then neighbourhood=borough
Replace 'Not assigned' with the value of that borough

```python
for x in df2['Neighbourhood']:
    if x=='Not assigned':
        df2['Neighbourhood']=df2['Borough'].where(df2['Neighbourhood'].eq('Not assigned'), df2['Neighbourhood'])
df2.head(10)
```

# Group the neighbourhood by the postcode and borough since some may have same postcode

```python
df3 = df2.groupby(['Postcode', 'Borough']).agg(', '.join).reset_index()
df3
```
